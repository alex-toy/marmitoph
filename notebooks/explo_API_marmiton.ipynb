## Function with Marmiton:
Renvoie une table de recettes Marmiton sur la base d'une liste d'ingrédients à inclure et à exclure

```python
import pandas as pd
from marmiton import Marmiton, RecipeNotFound

def get_recipes(ingredients_to_include = [], ingredients_to_exclude = []):

    # Creation of query keywords based on ingredients
    query_keywords = " ".join(ingredients_to_include)
    if len(ingredients_to_exclude) > 0:
        query_keywords += " sans "
        query_keywords += " ".join(ingredients_to_exclude)

    # Search :
    query_options = {"aqt": query_keywords,}
    query_result = Marmiton.search(query_options)

    # Recap on recipes and urls
    url_list = [recipe['url'] for recipe in query_result]
    recipes_list = [recipe['name'] for recipe in query_result]

    # Creation of tab with recipes infos
    recipes_keys = list(Marmiton.get(url_list[0]).keys())
    recipes_keys.remove('name')
    recipes_tab = pd.DataFrame(index = recipes_list, columns = ['url'] + recipes_keys)

    for url in url_list:
        recipe = Marmiton.get(url)
        recipes_tab.loc[recipe['name'], 'url'] = url
        for key in recipes_keys:
            recipes_tab.loc[recipe['name'], key] = recipe[key]
        
    return recipes_tab

```

```python
get_recipes(ingredients_to_include = ['boeuf', 'tomates', 'pates'], 
            ingredients_to_exclude = ['oignons']
            )
```

## Function with AllRecipes

```python
from allrecipes import AllRecipes
import pandas as pd

def get_recipes_AllRecipes(ingredients_to_include, ingredients_to_exclude = []):

    # Search :
    query_options = {
    "wt": "",         # Query keywords
    "ingIncl": ingredients_to_include,        # 'Must be included' ingrdients (optional)
    "ingExcl": " ".join(ingredients_to_exclude),  # 'Must not be included' ingredients (optional)
    "sort": "re"                # Sorting options : 're' for relevance, 'ra' for rating, 'p' for popular (optional)
    }
    query_result = AllRecipes.search(query_options)

    # Recap on recipes and urls
    url_list = [recipe['url'] for recipe in query_result]
    recipes_list = [recipe['name'] for recipe in query_result]

    # Creation of tab with recipes infos
    recipes_keys = list(AllRecipes.get(url_list[0]).keys())
    recipes_keys.remove('name')
    recipes_tab = pd.DataFrame(index = recipes_list, columns = ['url'] + recipes_keys)

    for url in url_list:
        recipe = AllRecipes.get(url)
        recipes_tab.loc[recipe['name'], 'url'] = url
        for key in recipes_keys:
            recipes_tab.loc[recipe['name'], key] = recipe[key]
        
    return recipes_tab
    

```
